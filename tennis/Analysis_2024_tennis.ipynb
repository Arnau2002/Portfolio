```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar solo los partidos de final ('F' en la columna 'round')
final_matches = matches_df[matches_df['round'] == 'F'].copy()

# Parte 1: Tendencia de la edad de los ganadores de torneos a lo largo de la temporada
plt.figure(figsize=(12, 6))
sns.scatterplot(data=final_matches, x='tourney_date', y='winner_age')
plt.title('Edad de los Ganadores de Torneos a lo Largo de la Temporada 2024')
plt.xlabel('Fecha del Torneo')
plt.ylabel('Edad del Ganador')
plt.grid(True)
plt.tight_layout()
plt.show()

# Podemos agregar una tendencia con una regresión lineal
plt.figure(figsize=(12, 6))
sns.regplot(data=final_matches, x=final_matches['tourney_date'].astype('int64') // 10**9, y='winner_age', scatter_kws={'alpha': 0.7})
plt.title('Tendencia de la Edad de los Ganadores de Torneos a lo Largo de la Temporada 2024')
plt.xlabel('Fecha del Torneo (timestamp)')
plt.ylabel('Edad del Ganador')
plt.grid(True)
plt.tight_layout()
plt.show()

# Parte 2: Edad de los ganadores de torneos en diferentes niveles de torneo
plt.figure(figsize=(10, 6))
sns.boxplot(data=final_matches, x='tourney_level', y='winner_age', order=final_matches['tourney_level'].value_counts().index)
plt.title('Edad de los Ganadores de Torneos por Nivel de Torneo en 2024')
plt.xlabel('Nivel del Torneo')
plt.ylabel('Edad del Ganador')
plt.grid(True)
plt.tight_layout()
plt.show()

# La categoría de Finals ( F ) cuenta Tour Finals y Next Gen Finals
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Calcular el ranking promedio de cada jugador durante el año
average_rankings = rankings_df.groupby('player_id')['rank'].mean().sort_values()

# Seleccionar los IDs de los 8 jugadores con el ranking promedio más bajo
top_8_average_players_ids = average_rankings.head(8).index

# Filtrar el DataFrame de rankings para incluir solo a estos 8 jugadores
top_8_evolution_consistent = rankings_df[rankings_df['player_id'].isin(top_8_average_players_ids)]

# Unir con players_df para obtener los nombres de los jugadores
top_8_evolution_consistent = pd.merge(top_8_evolution_consistent, players_df[['player_id', 'name_first', 'name_last']], on='player_id', how='left')
top_8_evolution_consistent['full_name'] = top_8_evolution_consistent['name_first'] + ' ' + top_8_evolution_consistent['name_last']

# Visualizar la evolución del ranking para estos 8 jugadores
plt.figure(figsize=(12, 8))
sns.lineplot(data=top_8_evolution_consistent, x='ranking_date', y='rank', hue='full_name', marker='o')
plt.title('Evolución del Ranking de los 8 Jugadores con Mejor Ranking Promedio en 2024')
plt.xlabel('Fecha del Ranking')
plt.ylabel('Ranking')
plt.gca().invert_yaxis()  # Invertir el eje Y para que el ranking 1 esté en la parte superior
plt.legend(title='Jugador')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convertir 'ranking_date' a datetime si aún no lo está
rankings_df['ranking_date'] = pd.to_datetime(rankings_df['ranking_date'], format='%Y%m%d')

# Encontrar la última fecha de ranking en 2024
end_date_2024 = rankings_df['ranking_date'].max()

# Definir un rango de ranking "cercano" al Top 8 al final del año
ranking_threshold_upper = 20
ranking_threshold_lower = 9  # Just outside the Top 8

# Filtrar a los jugadores que estaban dentro de este rango en la última fecha de ranking
final_contenders = rankings_df[rankings_df['ranking_date'] == end_date_2024]
final_contenders_ids = final_contenders[(final_contenders['rank'] >= ranking_threshold_lower) & (final_contenders['rank'] <= ranking_threshold_upper)]['player_id'].unique()

# Filtrar el DataFrame de rankings para incluir solo a estos jugadores
final_contenders_evolution = rankings_df[rankings_df['player_id'].isin(final_contenders_ids)]

# Unir con players_df para obtener los nombres de los jugadores
final_contenders_evolution = pd.merge(final_contenders_evolution, players_df[['player_id', 'name_first', 'name_last']], on='player_id', how='left')
final_contenders_evolution['full_name'] = final_contenders_evolution['name_first'] + ' ' + final_contenders_evolution['name_last']

# Visualizar la evolución del ranking para estos jugadores
plt.figure(figsize=(14, 9))
sns.lineplot(data=final_contenders_evolution, x='ranking_date', y='rank', hue='full_name', marker='o')
plt.title(f'Evolución del Ranking de Jugadores Cercanos al Top 8 a Final de 2024 (Rank {ranking_threshold_lower}-{ranking_threshold_upper} en {end_date_2024.strftime("%Y-%m-%d")})')
plt.xlabel('Fecha del Ranking')
plt.ylabel('Ranking')
plt.gca().invert_yaxis()  # Invertir el eje Y para que el ranking 1 esté en la parte superior
plt.legend(title='Jugador', bbox_to_anchor=(1.05, 1), loc='upper left') # Colocar la leyenda fuera del gráfico
plt.grid(True)
plt.tight_layout()
plt.show()
```

## players

```python
handedness_counts = players_df['hand'].value_counts()
print("\nDistribución de la handedness de los jugadores:")
print(handedness_counts)
```

```python

known_handedness_distribution = handedness_counts[handedness_counts.index != 'U']

print("Distribución de la handedness (sin Unknowns):")
print(known_handedness_distribution)
```

```python

known_handedness_distribution = known_handedness_distribution[known_handedness_distribution.index.isin(['R', 'L', 'A'])]

# Calcular el total de jugadores con handedness conocida
total_known_handedness = known_handedness_distribution.sum()

# Calcular el porcentaje de cada handedness
percentage_distribution = (known_handedness_distribution / total_known_handedness) * 100

print("Porcentaje de distribución de la handedness (sin Unknowns y posiblemente sin otras categorías):")
print(percentage_distribution)
```

```python

# Asumiendo que ya tienes el DataFrame 'players_df' cargado

# Filtrar para obtener jugadores con handedness conocida
players_df = players_df[players_df['hand'] != 'U']

# Calcular las 10 nacionalidades más frecuentes entre estos jugadores
nationality_counts = players_df['ioc'].value_counts().nlargest(10)

print("\nTop 10 nacionalidades de los jugadores:")
print(nationality_counts)
```

```python
# Obtener el ranking más reciente de cada jugador
latest_rankings = rankings_df.loc[rankings_df.groupby('player_id')['ranking_date'].idxmax()]

# Unir con la información de los jugadores
merged_df = pd.merge(latest_rankings, players_df, on='player_id', how='inner')

# Calcular el ranking promedio por handedness
average_ranking_by_hand = merged_df.groupby('hand')['rank'].mean().sort_values()
print("\nRanking promedio por handedness (ranking más reciente):")
print(average_ranking_by_hand)
```

```python
top_50_rankings = rankings_df[rankings_df['rank'] <= 50]
top_50_players_nationalities = pd.merge(top_50_rankings['player_id'], players_df[['player_id', 'ioc']], on='player_id', how='inner')['ioc'].value_counts().nlargest(10)
print("\nTop 10 nacionalidades con más jugadores en el Top 50 (último ranking registrado):")
print(top_50_players_nationalities)
```

```python
# Unir los DataFrames por el ID del ganador
winner_info = pd.merge(matches_df, players_df[['player_id', 'height']], left_on='winner_id', right_on='player_id', how='left')
winner_info = winner_info.rename(columns={'height': 'winner_height'})
winner_info = winner_info.drop(columns=['player_id'])

# Unir los DataFrames por el ID del perdedor
full_match_info = pd.merge(winner_info, players_df[['player_id', 'height']], left_on='loser_id', right_on='player_id', how='left')
full_match_info = full_match_info.rename(columns={'height': 'loser_height'})
full_match_info = full_match_info.drop(columns=['player_id'])

# Calcular la altura promedio de los ganadores por superficie
average_winner_height_by_surface = full_match_info.groupby('surface')['winner_height'].mean().sort_values(ascending=False)
print("\nAltura promedio de los ganadores por superficie:")
print(average_winner_height_by_surface)
```
