```python
liga = fotmob.fotmob_request('/leagues?id=87&ccode3=ESP')
```

```python
liga
```

```python

df_liga = pd.DataFrame(liga.json()['matches']['allMatches'])
df_liga
```

```python
df_liga[df_liga['round'] == '38']
```

```python
df_jornada = df_liga.iloc[:10]
```

```python
df_jornada
```

```python
ids_partidos = df_jornada.id.unique()
```

```python
from tqdm import tqdm
df_total = pd.DataFrame()
for partido in tqdm(ids_partidos):
    df = fotmob.get_match_shotmap(partido)
    df_total = pd.concat([df_total, df])
```

```python

df_total.situation.value_counts()
```

```python
penalties_df = df_total[df_total['situation'] == 'Penalty']
print(penalties_df)
```
