```python

# Aplanar columnas de múltiples niveles
season.columns = [' '.join(col).strip() for col in season.columns.values]

# Verifica cómo quedaron
print(season.columns.tolist())

```

```python
print(season.head())
```

```python
# Lista de nuevos nombres para las columnas "Unnamed"
new_names = ['Squad', 'Pl', 'Age', 'Poss']  # Continúa la lista según el número de columnas "Unnamed"

# Renombrar las columnas "Unnamed" con los nuevos nombres
season.columns = [
    new_names.pop(0) if col.startswith('Unnamed') else col
    for col in season.columns
]

# Verificar los cambios
print(season.columns.tolist())
```

 ## Descripción de las columnas
Squad: El nombre del equipo (por ejemplo, Arsenal, Aston Villa, etc.).

Pl: Número de jugadores disponibles en el equipo.

Age: Edad promedio de los jugadores del equipo.

Poss: Porcentaje promedio de posesión de balón por el equipo durante la temporada.

Playing Time MP: Número total de partidos jugados.

Playing Time Starts: Número de partidos en los que un jugador ha sido titular.

Playing Time Min: Total de minutos jugados en la temporada.

Playing Time 90s: Total de 90 minutos equivalentes jugados (equivalente a las veces que un jugador jugó un partido completo).

Performance Gls: Goles marcados por los jugadores del equipo.

Performance Ast: Asistencias de los jugadores del equipo.

Per 90 Minutes Gls: Promedio de goles por cada 90 minutos jugados.

Per 90 Minutes Ast: Promedio de asistencias por cada 90 minutos jugados.

Per 90 Minutes G+A: Promedio combinado de goles y asistencias por cada 90 minutos.

Per 90 Minutes G-PK: Promedio de goles por cada 90 minutos jugados, excluyendo los goles de penalti.

Per 90 Minutes xG: Promedio de "goles esperados" (xG, una métrica avanzada que calcula la calidad de las oportunidades de gol) por cada 90 minutos jugados.

Per 90 Minutes npxG: Promedio de "goles esperados no relacionados con penalti" por cada 90 minutos jugados.

Team ID: (un código hash o un identificador de base de datos para el equipo).
