## Plot ShotMap

```python
shots = fotmob.get_match_shotmap(4791867)
shots
```

```python
# Obtener los IDs únicos de los equipos
unique_team_ids = shots['teamId'].unique()
team_id_to_filter = unique_team_ids[0] # Seleccionamos el primero como ejemplo

# Filtrar el DataFrame de tiros por el teamId seleccionado
shots_filtered = shots[shots['teamId'] == team_id_to_filter].copy() # .copy() para evitar SettingWithCopyWarning

shots_filtered = shots_filtered.reset_index(drop=True)

print("DataFrame filtrado con índice reseteado:")
print(shots_filtered)
```

#### Quitamos los tiros de la tanda de penaltis final

```python
shots_filtered = shots_filtered[['x', 'y']]
shots_filtered = shots_filtered[0:9]

shots_filtered
```

```python


pitch = Pitch(pitch_type='uefa', # Uefa’s standard (105m * 65m).
              axis=True, label=True,            # Mostrar ejes y etiquetas es útil
              half=True,                       # Si quieres el campo completo o solo una mitad
              goal_type='box',                  # Tipo de portería ('box' o 'line')
              line_color='white',               # Color de las líneas del campo
              pitch_color='mediumseagreen'      # Color del césped
             )
fig, ax = pitch.draw(figsize=(10, 7)) # Puedes ajustar el tamaño de la figura aquí

# Dibujar los puntos de los tiros
# pitch.scatter() es el método preferido de mplsoccer para plotear puntos en el campo
scatter_shots = pitch.scatter(shots_filtered['x'], shots_filtered['y'],  # Coordenadas X e Y de tus tiros
                              s=180,                     # Tamaño de los puntos
                              color='red',               # Color de los puntos
                              alpha=0.7,                 # Transparencia de los puntos
                              edgecolors='black',        # Color del borde de los puntos
                              ax=ax,                     # El objeto Axes donde dibujar
                              label='Tiros')    # Etiqueta para la leyenda

# Añadir un título
ax.set_title('Mapa de Tiros en el Partido del Visitante', fontsize=20)

# Añadir leyenda si tienes múltiples tipos de puntos (por ejemplo, goles vs. fallos)
ax.legend(loc='lower left', fontsize=14)

plt.show()
```
